```python
from pyDes import *
```

```python
message = "0123456701234567"
key = "DECRYPTs"
iv = bytes([1]*8)
k = des(key, ECB, iv, pad = None ,padmode=PAD_PKCS5)
k2 = des(key, CBC, iv, pad = None ,padmode=PAD_PKCS5)
```

```python
cipher = k.encrypt(message)
cipher2 = k2.encrypt(message)

print("Len of message text: ", len(message))
print("Len of cipher text: ", len(cipher))
print()
print("ECB STYLE")
print()
print("Encrypted: ", cipher)
print("Encrypted: ", cipher[0:8])
print("Encrypted: ", cipher[8:16]) #bloques id√©nticos ECB
print("Encrypted: ", cipher[16:])
print()
print("Len of message text: ", len(message))
print("Len of cipher text: ", len(cipher2))
print()
print("CBC STYLE")
print()
print("Encrypted: ", cipher2)
print("Encrypted: ", cipher2[0:8])
print("Encrypted: ", cipher2[8:16]) #bloques distintos CBC
print("Encrypted: ", cipher2[16:])
```
