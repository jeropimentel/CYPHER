```python
def generate_key(n):
    letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789'
    key = {}
    count = 0
    for c in letters:
        key[c] = letters[(count + n) % len(letters)]
        count += 1
    return key
```

```python
def encrypt(key, message):
    cipher = ""
    for c in message:
        if c in key:
            cipher += key[c]
        else:
            cipher += c
    return cipher

def decypher(key, message):
    decypher = "" 
    for c in message:
        if c in key:
            for k,v in key.items():
                if c == v:
                    decypher = decypher + k
        else:
            decypher = decypher + c
    return decypher
```

```python
key = generate_key(4)
message = "Aunquex sean una mierda, los voy a votar: 100 a単os de decadencia argentina resumidos en una frase. No aprendemos mas."
encrypt(key, message)
```
