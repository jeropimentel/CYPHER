```python
from pyDes import *


def modify(cipher):
    mod = [0]*len(cipher)
    #mod[10] = ord(' ') ^ ord('1')
    #mod[11] = ord(' ') ^ ord('0')
    #mod[12] = ord('1') ^ ord('0')
    mod[8] = 1
    return bytes([mod[i]^cipher[i] for i in range(len(cipher))])



message = "Give Bob"
key = "DESCRYPT"
iv = bytes([0]*8)
k = des(key, ECB, iv, pad = None ,padmode=PAD_PKCS5)
```

```python
cipher = k.encrypt(message)
print("Length: ", len(message))
print("Length: ", len(cipher))
print("Encrypt: ", cipher)
message = k.decrypt(cipher)
message
                    
```
