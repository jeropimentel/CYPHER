```python
#ENTROPY and CRYPTO
#pseudo-random
import random

class KeyStream():
    def __init__(self, key=1):
        self.next = key
    
    def rand(self):
        self.next = (1103515245*self.next + 12345)% 2**31
        return self.next
    
    def get_key_bite(self):
        return (self.rand()//2**23)% 255 # hago que el key tenga 31 bits
    
def encrypt(key, message):
    return bytes([message[i]^key.get_key_bite() for i in range(len(message))])
    
def transmit(cipher, likely):
    b = []
    for c in cipher:
        if random.randrange(0, likely) < 0.5:
            c = c^ 2**random.randrange(0,8)
        b.append(c)
    return bytes(b)
    
def modification(cipher):
    mod = [0]*len(cipher)
    mod[10] = ord(' ') ^ ord('1')
    mod[11] = ord(' ') ^ ord('0')
    mod[12] = ord('1') ^ ord('0')
    return bytes([mod[i]^cipher[i] for i in range(len(cipher))])

def get_key(message, cipher):
    return bytes([message[i]^cipher[i] for i in range(len(message))])

def crack(key_stream, cipher):
    lenght = min(len(key_stream), len(cipher))
    return bytes([key_stream[i]^cipher[i] for i in range(lenght)])

def brute_force(plain, cipher):
    for k in range(2**31):
        bf_key = KeyStream(k)
        for i in range(len(plain)):
            xor_value = plain[i]^cipher[i]
            if xor_value != bf_key.get_key_bite():
                break
        else:
            return k
    return False
    
    
```
