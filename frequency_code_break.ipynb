```python
def generate_key(n):
    letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    key = {}
    cnt = 0
    for c in letters:
        key[c] = letters[(cnt + n) % len(letters)]
        cnt += 1
    return key


def get_decryption_key(key):
    dkey = {}
    for c in key:
        dkey[key[c]] = c
    return dkey

    
def encrypt(key, message):
    cipher = ""
    for c in message:
        if c in key:
            cipher += key[c]
        else:
            cipher += c
    return cipher


# this is done by your enemy
key = generate_key(3)
message = "YOU ARE AWESOME"
cipher = encrypt(key, message)

# this is us breaking the cipher
for i in range(26):
    dkey = generate_key(i)
    cipher = encrypt(dkey, message)
    print(cipher)
```

```python
def faculty(n):
    if n <= 1:
        return n
    else:
        return faculty(n-1)*n
    
def counter(n):
    count = 0
    for i in range(n):
        count += 1
    return count
```

```python
import random
def generate_random_key():
    letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
    cletters = list(letters)
    key = {}
    for c in letters:
        key[c] = cletters.pop(random.randint(0, len(cletters)-1))
    return key
```
